```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data=pd.read_csv(r'C:\Users\Dell\Desktop\Priyanshi_Python_notepade_jupeter_work\Statistics_Python\Exploratory_Data_Analysis_Project\Project_1\hotel_bookings.csv')
pd.set_option("display.max_column",None)
Data
```

```python
df=Data.copy()
```

```python
df.isnull().sum().sort_values(ascending=False)[:]
```

```python
df=df.drop(df[(df.adults+df.babies+df.children)==0].index)
```

```python
df['company']=df['company'].fillna(0)
df['agent']=df['agent'].fillna(0)
```

```python
df['country'].fillna(df.country.mode().to_string(),inplace=True)
```

```python
df['children'].fillna(round(df.children.mean()),inplace=True)
```

```python
df.info()
```

```python
df['children']=df['children'].astype('int64')
df['agent']=df['agent'].astype('int64')
df['company']=df['company'].astype('int64')
df['adr']=df['adr'].astype('int64')
```

```python
df.info()
```

```python
# 1. How Many Booking Were Cancelled :

df['is_canceled'].value_counts()
```

```python
# 1.1 by chart :

df['is_canceled'].value_counts().plot(kind='bar')
```

```python
# 2. What is the booking ratio between Resort Hotel and City Hotel? :

df_not_canceled=df[df['is_canceled']==0]
```

```python
Series=df_not_canceled['hotel'].value_counts()
Series
```

```python
Ratio=round(Series/Series.sum()*100)
Ratio
```

```python
# 2.1 by chart:

sns.barplot(x=Ratio.index,y=Ratio.values,data=Ratio)
plt.xlabel("hotel")
```

```python
# 3. What is the percentage of booking for each year? :

df_not_canceled['arrival_date_year'].value_counts()
```

```python
yr_data=df_not_canceled['arrival_date_year'].value_counts()
yr=round(yr_data/yr_data.sum()*100)
yr
```

```python
# 3.1 by Chart :

sns.barplot(x=yr.index,y=yr.values,data=yr)
```

```python
# 3.2 by chart :

sns.countplot(x='arrival_date_year', hue='hotel', data=df_not_canceled)
```

```python
# 4. Which is the most busy month for hotel? :

df_not_canceled['arrival_date_month'].value_counts()
```

```python
new_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sorted_months =df_not_canceled['arrival_date_month'].value_counts().reindex(new_order)
sorted_months
```

```python
# 4.1 by chart :

plt.figure(figsize=(18, 6))
sorted_data = sorted_months/sorted_months.sum()*100
sns.lineplot(x=sorted_data.index,y=sorted_data.values,data=sorted_data)
```

```python
new_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

## Select only City Hotel :
sorted_months = df_not_canceled.loc[df.hotel=='City Hotel','arrival_date_month'].value_counts().reindex(new_order)
city_data=sorted_months/sorted_months.sum()*100
city_data
```

```python
new_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

## Select only Resort Hotel :
sorted_months = df_not_canceled.loc[df.hotel=='Resort Hotel','arrival_date_month'].value_counts().reindex(new_order)
resort_data = sorted_months/sorted_months.sum()*100
resort_data
```

```python
# 4.2 by chart :

plt.subplots(figsize=(18,6))
sns.lineplot(x=city_data.index,y=city_data.values,data=city_data,label='city Hotel')
sns.lineplot(x=resort_data.index,y=resort_data.values,data=resort_data,label='resort Hotel')
plt.xlabel('Months')
plt.ylabel('Booking (%)')
```

```python
# 6. How Long People Stay in the hotel? :
total_nights = df_not_canceled['stays_in_weekend_nights']+df_not_canceled['stays_in_week_nights']
total_data=total_nights.value_counts()[0:10]
total_data
```

```python
total_night_data=total_data/total_data.sum()*100
total_night_data
```

```python
# 6.1 by chart :
# sns.barplot(x=total_night_data.index,y=total_night_data.values,data=total_night_data)

plt.bar(total_night_data.index,total_night_data.values)
```

```python
df_not_canceled.loc[:,'total_nights'] = df_not_canceled['stays_in_weekend_nights']+df_not_canceled['stays_in_week_nights']
fig, ax = plt.subplots(figsize=(12,6))
ax.set_xlabel('No of Nights')
ax.set_ylabel('No of Nights')
ax.set_title('Hotel wise night stay duration (Top 10)')
sns.countplot(x='total_nights', hue='hotel', data=df_not_canceled,
order =df_not_canceled.total_nights.value_counts().iloc[:10].index, ax=ax)
```

```python
# 7.Which was the most booked accommodation type (Single, Couple, Family) :

## Select single, couple, multiple adults and family
single = df_not_canceled[(df_not_canceled.adults==1) & (df_not_canceled.children==0) & (df_not_canceled.babies==0)]
couple = df_not_canceled[(df_not_canceled.adults==2) & (df_not_canceled.children==0) & (df_not_canceled.babies==0)]
family = df_not_canceled[df_not_canceled.adults + df_not_canceled.children + df_not_canceled.babies > 2]

## Make the list of Category names, and their total percentage
names = ['Single', 'Couple (No Children)', 'Family / Friends']
count = [single.shape[0],couple.shape[0], family.shape[0]]
count_percent = [x/df_not_canceled.shape[0]*100 for x in count]
```

```python
# 7.1 by chart :

sns.barplot(x=names,y=count_percent)
```
